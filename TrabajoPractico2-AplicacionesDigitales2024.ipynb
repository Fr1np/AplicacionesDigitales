## Ejercicio 9
Escribir una función que calcule el máximo común divisor de dos números y otra que calcule el mínimo común múltiplo.


````python
def maxComDiv(num1,num2):
    while num2 != 0:
        num1, num2 = num2, num1%num2
    return num1
def minComMul(num1, num2):
    return abs(num1 * num2) // maxComDiv(num1, num2)
print(maxComDiv(6,254), minComMul(4,7))
````

````python
def descuento(precio,desc):
    return precio-(precio*desc)

def IVA(precio,iva):
    return precio + (precio*iva)

def canasta(dicc, oper):
    lista_precios = []
    if oper == 'DESCUENTO':
        for i in dicc:
            entrada = [i, descuento(
                dicc[i]['Precio'],
                float('0.'+str(dicc[i]['Descuento']))
            )]
            lista_precios.append(entrada)
    elif oper == 'IVA':
        for i in dicc:
            entrada = [i, IVA(
                dicc[i]['Precio'],
                float('0.'+str(dicc[i]['iva']))
            )]
            lista_precios.append(entrada)
    return f'Precio con {oper.capitalize()} -->', lista_precios

diccionario={
    'Queso':{'Precio':45,'Descuento':45,'iva':21},
    'Pure':{'Precio':20,'Descuento':70, 'iva':16},
    'Tomate':{'Precio':37, 'Descuento':25, 'iva':6}
}
canasta(diccionario, input('').upper())
````

## Ejercicio 2
Escribir una función que simule una calculadora científica que permita calcular el seno, coseno, tangente, exponencial y logaritmo neperiano. La función preguntará al usuario el valor y la función a aplicar, y mostrará por pantalla una tabla con los enteros de 1 al valor introducido y el resultado de aplicar la función a esos enteros.


````python
import math

def seno(nums):
    resultado={}
    for i in nums:
        resultado[i]=math.sin(i)
    return resultado
def coseno(nums):
    resultado={}
    for i in nums:
        resultado[i]=math.cos(i)
    return resultado
def tangente(nums):
    resultado={}
    for i in nums:
        resultado[i]=math.tan(i)
    return resultado
def exponencial(nums):
    resultado={}
    for i in nums:
        resultado[i]=math.exp(i)
    return resultado
def logaritmoNeperiano(nums):
    resultado={}
    for i in nums:
        resultado[i]=math.log(i)
    return resultado

def calculadora(num,operacion):
    lista_nums = list(range(1,num+1))
    if operacion == 'seno':
        resultado = seno(lista_nums)
        print('Resultado del seno de los numeros')
        for i in resultado:
            print(f'{i} = {resultado[i]}')
    elif operacion == 'coseno':
        resultado = coseno(lista_nums)
        print('Resultado del coseno de los numeros')
        for i in resultado:
            print(f'{i} = {resultado[i]}')
    elif operacion == 'tangente':
        resultado = tangente(lista_nums)
        print('Resultado de la tangente de los numeros')
        for i in resultado:
            print(f'{i} = {resultado[i]}')
    elif operacion == 'exponencial':
        resultado = exponencial(lista_nums)
        print('Resultado del exponencial de los numeros')
        for i in resultado:
            print(f'{i} = {resultado[i]}')
    elif operacion == 'logaritmo' or operacion=='logaritmo neperiano':
        resultado = logaritmoNeperiano(lista_nums)
        print('Resultado del logaritmo neperiano de los numeros')
        for i in resultado:
            print(f'{i} = {resultado[i]}')
            
calculadora(int(input()), input().lower())
    
````

## Ejercicio 4

## Ejercicio 8
Escribir una función que calcule el módulo de un vector.


````python
import math
def modVec(vector):
    cuadrados = sum(componente ** 2 for componente in vector)
    modulo = math.sqrt(cuadrados)
    return modulo

vector = [3, 4, 5, 6]
print(f'El módulo del vector {vector} es: {modVec(vector)}')
````

## Ejercicio 9
Una inmobiliaria de una ciudad maneja una lista de inmuebles como la siguiente:

[{'año': 2000, 'metros': 100, 'habitaciones': 3, 'garaje': True, 'zona': 'A'},
{'año': 2012, 'metros': 60, 'habitaciones': 2, 'garaje': True, 'zona': 'B'},
{'año': 1980, 'metros': 120, 'habitaciones': 4, 'garaje': False, 'zona': 'A'},
{'año': 2005, 'metros': 75, 'habitaciones': 3, 'garaje': True, 'zona': 'B'},
{'año': 2015, 'metros': 90, 'habitaciones': 2, 'garaje': False, 'zona': 'A'}]

Construir una función que permita hacer búsqueda de inmuebles en función de un presupuesto dado. La función recibirá como entrada la lista de inmuebles y un precio, y devolverá otra lista con los inmuebles cuyo precio sea menor o igual que el dado. Los inmuebles de la lista que se devuelva deben incorporar un nuevo par a cada diccionario con el precio del inmueble, donde el precio de un inmueble se calcula con las siguiente fórmula en función de la zona:

Zona A: precio = (metros * 1000 + habitaciones * 5000 + garaje * 15000) * (1-antiguedad/100)

Zona B: precio = (metros * 1000 + habitaciones * 5000 + garaje * 15000) * (1-antiguedad/100) * 1.5


````python
def precios(lista):
    for i in range(len(lista)):
        if lista[i]['zona']=='A':
            antiguedad = 2024 - lista[i]['año']
            lista[i]['precio'] = round((lista[i]['metros']*1000 + lista[i]['habitaciones']*5000 + lista[i]['garaje']*15000)*(1-antiguedad/100))
        if lista[i]['zona']=='B':
            antiguedad = 2024 - lista[i]['año']
            lista[i]['precio'] = round((lista[i]['metros']*1000 + lista[i]['habitaciones']*5000 + lista[i]['garaje']*15000)*(1-antiguedad/100) * 1.5)

def busqueda(lista, entrada):
    lista_opciones = []
    for i in range(len(lista)):
        if lista[i]['precio'] <= entrada:
            lista_opciones.append(lista[i])
        else:
            continue
    return lista_opciones

lista_inmuebles = [{'año': 2000, 'metros': 100, 'habitaciones': 3, 'garaje': True, 'zona': 'A'},
                   {'año': 2012, 'metros': 60, 'habitaciones': 2, 'garaje': True, 'zona': 'B'},
                   {'año': 1980, 'metros': 120, 'habitaciones': 4, 'garaje': False, 'zona': 'A'},
                   {'año': 2005, 'metros': 75, 'habitaciones': 3, 'garaje': True, 'zona': 'B'},
                   {'año': 2015, 'metros': 90, 'habitaciones': 2, 'garaje': False, 'zona': 'A'}]

precios(lista_inmuebles)
busqueda(lista_inmuebles, int(input('Precio deseado: ')))
````

## Ejercicio 10
Escribir una función que reciba una muestra de números y devuelva los valores atípicos, es decir, los valores cuya puntuación típica sea mayor que 3 o menor que -3. Nota: La puntuación típica de un valor se obtiene restando la media y dividiendo por la desviación típica de la muestra.

````python
import numpy as np

def valoresAtipicos(muestra):
    media = np.mean(muestra)
    desviacion_estandar = np.std(muestra)

    valores_atipicos = []
    for valor in muestra:
        puntuacion_tipica = (valor - media) / desviacion_estandar
        if abs(puntuacion_tipica) > 3:
            valores_atipicos.append(valor)

    return valores_atipicos


muestra = [1, 2, 3, 4, 5, 100, 200, -10, -20]
valores_atipicos = valoresAtipicos(muestra)
print(valores_atipicos)
````
